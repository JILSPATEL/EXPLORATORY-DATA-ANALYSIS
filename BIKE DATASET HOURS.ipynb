```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

# 2. Reading and Understanding the data

```python
df2=pd.read_csv('hour.csv')
```

```python
df2.head(2)
```

# 3.Cleaning data

```python
df2.drop(['instant','dteday'],axis=1,inplace=True)
```

```python
df2.drop(['holiday'],axis=1,inplace=True)
```

```python
df2.rename(columns={'yr':'year','mnth':'month','temp':'temprature','hum':'humidity','hr':'hours','cnt':'count'},inplace=True)
```

```python
df2.head(2)
```

```python
season_code={1:'spring',2:'summer',3:'fail',4:'winter'}
df2['season']=df2['season'].map(season_code)
```

```python
weather_code={1:'clear',2:'mist',3:'light rain',4:'heavy rain'}
df2['weathersit']=df2['weathersit'].map(weather_code)
```

```python
working_codes = {1:'working_day',0:'Holiday'}
df2['workingday'] = df2['workingday'].map(working_codes)
```

```python
month_codes = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
df2['month']=df2['month'].map(month_codes)
```

```python
weekday_codes = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df2['weekday'] = df2['weekday'].map(weekday_codes)
```

```python
year_code={0:"2018",1:"2019"}
df2['year']=df2['year'].map(year_code)
```

```python
df2.head(5)
```

# 4. Dividing column into continuous and categorical column

```python
constant=[ 'temprature','atemp','humidity','windspeed','casual','registered']
categorical=['season','year','month','weekday','workingday','weathersit']
tg = ['count']
```

# 5. Univaraite Analysis

```python
for i in constant:
    sns.distplot(df2[i])
    plt.show()
```

```python
for i in categorical:
    sns.countplot(df2[i])
    plt.show()
```

# 6. Bivariate Analysis

```python
for data in constant:
    sns.scatterplot(df2[data],df2['count'],hue=df2['workingday'])
    plt.show()
```

```python
for data in constant:
    sns.scatterplot(df2[data],df2['count'],hue=df2['year'])
    plt.show()
```

```python
for data in categorical:
    sns.boxplot(df2[data],df2['count'])
    plt.show()
```

```python
for data in categorical:
    sns.barplot(df2[data],df2['count'])
    plt.show()
```

# 7. Plotting heatmap

```python
plt.figure(figsize=(14,8))
sns.heatmap(df2.corr(),annot=True,cmap='Reds')
plt.show()
```
